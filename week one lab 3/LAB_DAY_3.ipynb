```python
data["temperature_C"].fillna(data["temperature_C"].mean(), inplace=True)
data["humidity_%"].fillna(data["humidity_%"].mean(), inplace=True)
data["wind_speed_kmph"].fillna(data["wind_speed_kmph"].mean(), inplace=True)
data["pressure_hPa"].fillna(data["pressure_hPa"].mean(), inplace=True)
data["visibility_km"].fillna(data["visibility_km"].mean(), inplace=True)
```

```python

data["hour"]=data["timestamp"].dt.hour
avg_humidity_by_hour = data.groupby("hour")["humidity_%"].mean()

most_humid_hour = avg_humidity_by_hour.idxmax()
highest_humidity = avg_humidity_by_hour.max()

print("\nAverage Humidity by Hour:")
print(avg_humidity_by_hour)

print(f"\nMost humid hour: {most_humid_hour}")
print(f"Average Humidity: {highest_humidity}")
```

```python
temp_matrix = temp.reshape((10,24))
print(temp_matrix)
print("minimum",temp_matrix.min(axis=1))
print("maximum",temp_matrix.max(axis=1))
print("mean",temp_matrix.mean(axis=1))
```

```python
def normalize(matrix):
    mean=np.mean(matrix)
    std=np.std(matrix)
    return (matrix-mean)/std

print(normalize(temp_matrix))

```

```python
mask = wind > 15

high_wind = wind[mask]

print(high_wind)
```

final challange 

```python

def daily_summary(matrix):
    summaries = []
    for i in range(10):
        day_summary = {
            "Day": i + 1,
            "Minimum": float(np.min(matrix[i])),
            "Maximum": float(np.max(matrix[i])),
            "Mean": float(np.mean(matrix[i]))
        }

        summaries.append(day_summary)

    return summaries


summaris = daily_summary(temp_matrix)
for day in summaris:
    print(day)
```
