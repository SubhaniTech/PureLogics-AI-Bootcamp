```python

import pandas as pd

pakistani_weather = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
pakistani_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df=pd.concat([pakistani_weather, us_weather])
df
```

```python
df1 = pd.concat([pakistani_weather, us_weather], ignore_index=True)
df1
```

```python
df3 = pd.concat([pakistani_weather, us_weather], keys=["Pakistan", "US"])
df3
```

```python
df3.loc["US"]
```

```python
df3.loc["Pakistan"]
```

```python
temperature_df = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["lahore","karachi"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df_new=pd.concat([temperature_df, windspeed_df], axis=1)
df_new
```
