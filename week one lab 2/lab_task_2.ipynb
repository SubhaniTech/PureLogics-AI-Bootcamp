STEP 5 : TUPLE , SETS , DICTIONARY

```python
# perform basic operations on tuples
tuple_1 = (1, 2, 3, 4, 5)
tuple_2 = (6, 7, 8, 9, 10)
# Concatenation
added_tuple = tuple_1 + tuple_2
print("Concatenated tuple:", added_tuple)

# find the length of tuple by using len function 
print(f"length of string  in added tuple ",len(added_tuple))

#find the max and min value in the tuple
print("max value:",max(added_tuple))
print("min value:",min(added_tuple))

#slicing of tuple 
sliced_tuple =(2,4,2,1,5,1)  
print("sliced_tuple",sliced_tuple[2:5])# Slicing from index 2 to 5

#used negitive indexing in tuple 
neg=[12,2,3,4,4,5,6,7,8,9,10]
print("negitive indexing in tuple",neg[-1])

#used in or not in operators
x=("banana","mango","apple")
print("banana" in x)  # True
print("mango" not in x)  # True


```

```python
#operation perform on dictionary
my_dict = {'name': 'huzaifa', 'age': 23, 'city': 'lahore'}
print(my_dict)

#accessing values in dictionary by using index number 
print(my_dict["name"])

#accessing values in dictionary by using get() method
print(my_dict.get("age"))

#adding a new key in a dictionary
my_dict["country"] = "Pakistan"
print(my_dict)

#removing an item from a dictionary by using pop() method
my_dict.pop("age")
print(my_dict)

#removing an item from a dictionary by using del() method
del my_dict["city"]
print(my_dict)

#find length of dictionary by using len() method
print("\nlength ",len(my_dict))

#get all the keys in a dictionary by using keys() method
print("\n keys ",my_dict.keys())

#get all the values in a dictionary by using values() method
print("\n values ",my_dict.values())

#get all key values pairs in a dictionary by using items() method
print("\n items ",my_dict.items())
```

```python
#perform operation on sets
set_1 = set() #create an empty sets
set_1.add(1)
set_1.add(2)
set_1.add(3)
print(set_1)

#update a sets
set_1.update([2,3,4,5])
print("updatesd set ",set_1)

#remove an items
set_1.remove(2)
print(":after removing an item",set_1)

#remove an element by using discard function
set_1.discard(3)
print("after discarding an item",set_1)

#find the length of sets by using len() method
print(": the length of sets", len(set_1))
```

STEP 6 :  WORD FREQUENCY COUNTER

```python
one ="Python is simple yet powerful. Python is easy to learn."
one=one.replace(".","")
text=one.split()#split the string into words
print("total words in the string",len(text))

word_freq = {}
print(word_freq)
# Count word frequency
for word in text:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

print(word_freq)




```
